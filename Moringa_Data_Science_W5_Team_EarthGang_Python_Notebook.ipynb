# Data Cleaning

Lets clean our primary school dataset

### Validity

```
# Checking for similarity of columns
df[ ~(df.X == df['Longitude'])]
```

```
# Dropping unnecesary columns
prim = df.drop(['X','Y','Level_','ClassrmRat','PupilTeach',\
         'GO_KTSC_M','GOK_TSC_F','TotalBoys','TotalGirls','authorityM','authorityF',\
         'PTA_BOG_M','PTA_BOG_F','OthersM','OthersF','NoTeaching','NonTeachin',\
         'Division','Latitude','Longitude', 'SchSponsor'], axis = 1)
prim.head()
```

### Completeness

```
# Check for null values
prim.isnull().sum()
```

```
# Dropping null values
prim_1 = prim.dropna()
```

### Consistency

```
# Checking for duplicates
prim_1.duplicated().sum()
```

### Uniformity

```
pd.options.mode.chained_assignment = None  
```

```
# Changing the case of column names to lower, stripping whitespaces,
# removing periods and replacing spaces with underscore.
prim_1.columns = prim_1.columns.str.lower().str.strip().str.replace(' ', '_')
prim_1.head()
```

```
# Renaming columns to have more relevant names
prim_1.rename(columns = {'name_of_sc':'school_name', 'schsponsor': 'school_sponsor','totaltoile':'total_toilets',\
                         'toiletrati':'toilet_ratio','totalenrol':'total_students','teachersto':'teachers_toilet',\
                         'costituenc':'constituency'}, inplace = True)
prim_1.head()
```

```
# Filling totaltoilet column with sum of boystoilet and girlstoilet
toilets = prim_1['boystoilet']+prim_1['girlstoilet']
prim_1['total_toilets']= toilets
prim_1
```

```
# Renaming of Central province to CENTRAL for uniformity.
prim_1['province'].replace({'Central':'CENTRAL'},inplace=True)
```

### Merging our datasets

Now that we have cleaned up our primary school dataset, lets merge it to the counties dataset.

```
# Merging our two datasets
new_df = prim_1.merge(counties, on= ['district'], how= 'outer')
new_df
```

### Cleaning after Merging

```
# Checking for null values in our merged dataframe
new_df.isnull().sum()
```

```
# Replacing null values with zeros
new_df.fillna(0)
new_df[new_df.fid != '31229']
new_df.dropna()
```

# Data Analysis

Analysis for this project will use the WHO recommended ratio of 1:27 which means one toilet to 27 students per school.

# Status of Sanitation in Provinces, Counties and District as per Student to Toilet Ratio



### What is the status of sanitation in the provinces as per Student to Toilet Ratio

```
# The ratio of students to toilets in each province.
prov_toilet=new_df.groupby('province')[['total_toilets']].sum()
prov_toilet
prov_stu=new_df.groupby('province')[['total_students']].sum()
prov_stu
prov_toilet_ratio=((prov_stu['total_students'])/(prov_toilet['total_toilets']))
print(prov_toilet_ratio)
```

```
# Provinces which do not meet WHO's threshold.
prov_toilet_ratio.loc[prov_toilet_ratio > 27.0]
```

7 out of 8 provinces do not meet WHO's threshold.

### What is the status of sanitation in the various counties as per Student to Toilet Ratio

```
# The ratio of students to toilets in each county.
county_toilet=new_df.groupby('county')[['total_toilets']].sum()
county_toilet
county_stu=new_df.groupby('county')[['total_students']].sum()
county_stu
county_toilet_ratio=((county_stu['total_students'])/(county_toilet['total_toilets']))
print(county_toilet_ratio)
```

```
# Counties which do not meet WHO's threshold.
print('Counties that do not meet WHO threshold are',county_toilet_ratio.loc[county_toilet_ratio>27.0].count())
county_toilet_ratio.loc[county_toilet_ratio > 27.0]
```

### What is the status of sanitation in districts as per Student to Toilet 

```
# The ratio of students to toilets in each district.
district_toilet=new_df.groupby('district')[['total_toilets']].sum()
district_toilet
district_stu=new_df.groupby('district')[['total_students']].sum()
district_stu
district_toilet_ratio=((district_stu['total_students'])/(district_toilet['total_toilets']))
district_toilet_ratio.tail()
```

### Comparison of status of sanitation between public and private schools

```
# Status of sanitation in Public Schools
public =new_df.loc[new_df['status']=='PUBLIC']
public
print("Total number of public schools =",public['fid'].count())

# Calculate total number of public schools without sanitation facilities
pub_1 = public.loc[(public['boystoilet']==0.0)&(public['girlstoilet']==0.0)&(public['teachers_toilet']==0.0)]
print("Total number of public schools without sanitation facilities =",pub_1['fid'].count())

# Calculating ratio of public shools to public schools without sanitation facilities
pub_ratio = float((pub_1['fid'].count())/(public['fid'].count()))
print("Percentage of public schools without sanitation =",(pub_ratio*100.0),"%")
print("The list of public schools without sanitation is shown below")
pub_1
```

```
# Status of sanitation in Private Schools
private = new_df.loc[new_df['status']=='PRIVATE']
private
print("Total number of private schools =",private['fid'].count())

# Calculate total number of public schools without sanitation facilities
priv_1 = private.loc[(private['boystoilet']==0.0)&(private['girlstoilet']==0.0)&(private['teachers_toilet']==0.0)]
priv_1['fid'].count()
print ("Total number of private schools without sanitation facilities =",priv_1['fid'].count())

# Calculating ratio of private schools to private schools without sanitation facilities
priv_ratio = float((priv_1['fid'].count())/(private['fid'].count()))
print("Percentage of private schools without sanitation =",(priv_ratio*100.0),"%")
print("The list of private schools without sanitation is shown below")
priv_1
```
